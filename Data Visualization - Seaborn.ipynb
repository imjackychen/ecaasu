**Modules:<br>
seaborn for statistical data visualization <br>
matplotlib for Python plotting**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
```

**load_dataset looks for datasets on** <br> https://github.com/mwaskom/seaborn-data

```python
titanic = sns.load_dataset('titanic')
```

**Displays the info() of the 'titanic' DataFrame**

```python
titanic.info()
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
```

```python
sns.distplot(titanic['fare'],bins=40,kde=False,color='green')
```

```python
sns.boxplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='Set2')
```

```python
sns.countplot(x='sex',data=titanic)
```

```python
sns.heatmap(titanic.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
g = sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
```
